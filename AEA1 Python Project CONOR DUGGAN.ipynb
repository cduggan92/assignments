### The Lewis model

*Note: the following discussion is based on [Ros (2000, 69-100)](https://books.google.nl/books/about/Development_Theory_and_the_Economics_of.html?id=ippZV3ZhdZUC&redir_esc=y)*

When we refer to a non capitalist or subsistence sector, we can think of it as being agricultural and employing neglibible levels of capital. There is also an abundance of available labour, such that the marginal product of labour is zero. The wage received for working in this sector is, however, not zero, as we assume that the product of labour is divided equally such that each worker receives the average product of their labour. An additional assumption is that workers there is no savings in the subsistence sector, with all income being consumed.

The following 7 simultaneous equations form the schedule for the short run equilibria of the Lewis model that I will simulate in Python below: 

#### Production functions

In the modern sector, capital is used with constant returns to scale, which we assume to be Cobb Douglas:

$$
Y_m = A_m K^\alpha L_m^{1-\alpha}
$$

Where $Y_m$ is the output of the modern sector, $A_m$ is the modern sector technology parameter, and $L_m$ is labour employed in the modern sector.

In the subsistence sector, output has constant returns to labour 

$$
Y_s = A_s * L_s
$$

Where $Y_s$ is the output of the subsitence sector, $A_s$ captures the lower state of technology in the substience sector, and $L_s$ is labour employed the sector.

#### Labour market and wages

Labor markets in this model are competitive in that “the wage which the expanding capitalist sector has to pay is determined by what people can earn outside that sector” ([(Lewis, 1954, 148)](http://onlinelibrary.wiley.com/doi/10.1111/j.1467-9957.1954.tb00021.x/abstract). In other words, the wage in capitalist sector has a premium, as the cost of living in urban areas is higher, living in industrial areas is considered less attractive, and the capitalist sector wishes to reduce turnover costs (efficiency wage theory). Ad the wage in the subsistence sector is equal to the average product of labour, the short run equilbirium is therefore simply:

$$
w_s = Y_s / L _s
$$

where $Y_s$ is the output of the subsistence sector and $L_s$ is labour employed in it.

The wage premium afforded to modern sector labour, $\phi$, is constant so that the wage payed by the modern sector is:

$$
w_m = \phi w_s
$$

Short run equilibirum employment in the capitalist sector is determined by typical profit maximisation given the level of technology. Given perfect competition in all markets, labour demand in the modern sector will be:

$$
L_m = K[ (1-\alpha)  A_m / w_m ] ^ {1 / \alpha}
$$

This holds as long as $L_m < L$ . Note that the short run equilibirum wage rate in the modern sector is independent of k and determined solely by the wage in subsistence sector and the wage premium. Therefore, as labour in the modern sector is increasig in capital employed, increases in the capital-labour ratio will have the affect of dragging labour into the modern sector, affecting the sectoral balance of labour but leaving equilibirum wage rates untouched.

A feature of the **Lewis** model is that there is no traditional unemployment, but rather *disguised* unemployment, whereby unused workers are "given" a job in the subsistence sector despite their MPL being 0. The total labour force is therefore either employed in the modern or subsistence sectors such that:

$$
L = L_m + L_s
$$

Where $L$ is the total labour force.

#### Capital accumulation

In the one-sector **Solow** model, capital accumulation takes place according to:

$$
    \dot K = sY - \delta K
$$

Where the change in capital is the proportion of output that is saved minus depreciation of the capital stock.

In the **Lewis** model however, it is not merely a fraction of output that is saved, but rather a fraction of output minus wages:

$$
    \dot K = s_m[Y_m - w_m L_m] - \delta K
$$

Note here that we assume workers in both sectors consume their entire wages, so there are no household savings, so capital accumulation occurs only through the reinvestment of capitalist surplus net of depreciation servicing.

#### Transition to mature economy

As we assume the capitalist follows classical profit maximisation, he will continue to employ labour up to the point where its marginal product equals the wage paid. The additional capital in the modern sector created from the capitalist surplus raises the marginal product of labour, as capital "fructifies" labour. The capitalist therefore demands more workers, but as labour is abundant in the subsistence sector, he can do so with no subsequent rise in modern sector wages. In this fashion the modern sector continues to absorb labour from the subsistence sector as capital accumlates while wages remain static. 

When the entirety of labour in the subsistence sector is eventually absorbed, the subsistence sector disappears, such that $$L_s = 0$$, at which point the economy has become a one-sector mature capitalist economy. The short run equilbirum of wages now becomes 

$$ 
w_m = A(1 - \alpha)k^\alpha
$$

This is identical to determination of wages the Solow model. From this point onwards, the model behaves exactly like the traditional [Solow (1956) model](https://www.econ.nyu.edu/user/debraj/Courses/Readings/Solow.pdf). I assume that the reader is familiar with the **Solow** model, and will not go in to it in detailhere for the sake of brevity. For a refresher, [Acemoglu's slides](https://economics.mit.edu/files/7181) on the topic provide a useful overview. 

In short, the key insights of the **Solow** model is that a one-sector capitalist economy will accumulate capital rapidly at the beginning, as these initial units of capital are very productive, creating high growth in output which results in high savings that are reinvested. However, diminishing returns set in, causing a lower increase in output for each additional unit of capital, resulting in a slowing level of investment and as such slowing accumulation of capital. 

Eventually, as depreciation is a constant function of capital, it will eventually equal investment, at which point we reach the *steady state level of capital and output*, where the entirety of savings is needed to replace depreciated capital. We will see these dynamics when we plot our simulation below. 

## Simulation

Using the above system of equations I now simulate the **Lewis** model with the end goal to visually graph the growth dynamics discussed above.

##### Calibration of exogenous parameters:
The model discussed above includes 7 exogenously defined parameters that we need to calibrate in order to run the model. These are:

1. $A_m$: state of technology in the modern sector
2. $A_s$:    state of technology in the subsistence sector
3. $L$:     total labour supply
4. $\alpha$: output elasticity of capital
5. $\phi$: wage premium in modern sector
6. $\delta$: capital depreciation
7. $s_\pi$: savings rate in modern sector

I begin by setting the states of technology in the modern and subsistence sectors such that the modern sector is significantly  more effective at getting output from given inputs, so I set $A_m = 2.65$ and $A_s=1.0$. 

Next I set the total labour supply, $L=1.0$, for simplicity, as it is not the level of labour that is important but the relative proportion of labour in each sector. 

I set the output elasticity of capital based on [Piketty's (2014)](http://piketty.pse.ens.fr/en/capital21c2) observance of capital's actual share of income across different countries, which is roughly 1/3, so $\alpha = 0.33$. 

Next I set the wage premium for working in the modern sector $\phi = 2.0 $, sufficiently high such that labour is enticed to switch sectors, and with a view of ensuring the dynamics of the model can be fully realised with a reasonable time horizon. 

The level of depreciation is set to $\delta = 0.3$, higher than that used by Piketty (2014) as capital in this model refers solely to productive machinery, with no housing capital as used by Piketty. 

Lastly, I set the savings rate in the modern sector, $s_\pi=0.3$. While this rate is somewhat [higher](http://1stclasseconomics.com/lewis-model/) than would be expected in a developing country and more suiting to a developed country, this rate is more conventially used in **Solow** modelling, and as such lends itself to the post-transition mature economy stage.

With the exogenous variables calibrated I can now begin my simulation.

### Importing packages

```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Define exogenous parameters

A_m = 2.65     #  state of technology in the modern sector
A_s = 1        #state of technology in the subsistence sector
L = 1          #total labour supply
alpha = 0.33   #output elasticity of capital
phi = 2        #wage premium in modern sector
delta = 0.1    #capital depreciation rate
s_m = 0.3      #savings rate in modern sector
K = 0.1        #Initial capital stock

one_minus_alpha = 1 - alpha

#Initialise empty lists for each variable

list_K = []
list_L_m = []
list_L_s = []
list_Y_s = []
list_Y_m = []
list_w_s = []
list_w_m = []
list_K_dot = []
list_Y = []

time_horizon = 100

for year in range(time_horizon):
    
    #Calculate endogenously define variables
    
    #Labour market equilibrium
    L_m = min((K* (( ( one_minus_alpha * A_m) / (phi * A_s) ) ** (1 / alpha))), 1)
    
    #Labour market
    L_s = L - L_m
     
    #Output in subsistence sector
    Y_s = A_s * L_s
    
    #Output in modern sector
    Y_m = A_m * (K ** alpha) * (L_m ** one_minus_alpha)
    
    #Wage in subsistence sector: making sure that L_m cannot exceed total labour, 1
    def wage_subsistence_sector(Y_s, L_s):
        if L_m < 1 :
            w_s = (Y_s / L_s)
            return w_s
        else:
            w_s = Y_s
            return w_s   
        
    w_s = wage_subsistence_sector(Y_s, L_s)
    
    #Wage in modern sector
    w_m = one_minus_alpha * (Y_m / L_m)
    
    #Capital accumulation
    K_dot = (s_m * (Y_m - w_m * L_m) ) - (delta * K)
    
    #Total output
    Y = Y_m + Y_s
    
    #Append endogenously define variables to lists
    list_K.append(K)
    list_L_m.append(L_m)
    list_L_s.append(L_s)
    list_Y_s.append(Y_s)
    list_Y_m.append(Y_m)
    list_w_s.append(w_s)
    list_w_m.append(w_m)
    list_K_dot.append(K_dot)
    list_Y.append(Y)
    
    #Update variables
    K = K + K_dot

#Put lists into pandas dataframe
time_series_df = pd.DataFrame( {'Capital' : list_K, 
                                'Labour Modern Sector' : list_L_m, 
                                'Labour Subsistence Sector' : list_L_s,
                                'Output Subsistence Sector' : list_Y_s,
                                'Output Modern Sector' : list_Y_m,
                                'Total Output' : list_Y,
                                'Wages Subsistence Sector' : list_w_s,
                                'Wages Modern Sector' : list_w_m,
                                'Change in Capital' : list_K_dot } )
time_series_df.index.name = 'Year'
```

```python
#Check if the dataframe we created makes sense
time_series_df
```

The table above displays the results of our simulation over 100 discrete time periods. We see the transition of labour from the subsistence sector to the modern sector, and the subsequent growth dynamics as a mature economy. While displaying the dataframe like this is somewhat useful, it is difficult to get to visualise the growth dynamics. I now turn to plotting these dynamics to aid in our understanding below. 

## Plotting the growth dynamics of the Lewis model

### Capital Accumulation

```python
#Plot capital accumulation
plt.clf()
plt.style.use('seaborn')
plt.plot(time_series_df['Capital'], label = 'Capital stock')
plt.plot(time_series_df['Change in Capital'], label = 'Change in Capital')
plt.axvline(27, lw=1, linestyle='dashed')
plt.text(28, 4.15, 'Mature economy', fontsize=9)
plt.xlabel('Year')
plt.ylabel('Capital')
plt.title('Figure 1 - Capital accumulation', fontweight='bold')
plt.legend()
plt.show()
```

**Figure 1**  shows the dynamics of per period change in capital stock, which we called $\dot K$ above, and the level of the capital stock. We see the change in capital stock accelerating during the dual-sector phase of the figure, before the dotted line marked *Mature economy*. Here, the capitalist surplus net of depreciation servicing is reinvested in additional capital. This additional capital raises the marginal productivity of labour in the modern sector, drawing labour into it as we will see in the next figure.

The line segments to the left of the dotted line in **Figure 1** show the dynamics of capital in the mature one-sector economy, which behaves exactly like a **Solow** model. Here we see capital accumulate quickly in the initial stages, but the realities of diminishing returns now set in due to the now finite labour available, making each unit of capital less productive than the last. Eventually, capital accumulation levels off, as savings are consumed entirely by replacing the existing capital stock that has depreciated.

### Sectoral movement of labour

```python
#Plot labour by sector
plt.clf()
plt.style.use('seaborn')
plt.plot(time_series_df['Labour Subsistence Sector'], label = 'Subsistence Sector')
plt.plot(time_series_df['Labour Modern Sector'], label = 'Modern Sector')
plt.axvline(27, lw=1, linestyle='dashed')
plt.xlabel('Year')
plt.ylabel('Proportion of workers')
plt.title('Figure 2 Labour employed by sector', fontweight='bold')
plt.legend()
plt.show()
```

**Figure 2** shows the migration of labour from the subsistence sector to the modern sector, as demand for labour is increased due to the higher MPL resulting from capital accumulation. When the labour employed in the subsistence sector reaches 0 the economy has become a one-sector modern economy.

### Wages

```python
#Plot wages by sector
plt.clf()
plt.style.use('seaborn')
plt.plot(time_series_df['Wages Subsistence Sector'], label = 'Subsistence Sector')
plt.plot(time_series_df['Wages Modern Sector'], label = 'Modern Sector')
plt.axvline(27, lw=1, linestyle='dashed')
plt.xlabel('Year')
plt.ylabel('Wage')
plt.title('Figure 3 - Wages by sector', fontweight='bold')
plt.legend()
plt.show()
```

**Figure 3** above shows graphical evidence of the static wages resulting from labour abundance as discussed previously. Wages begin to rise in the mature economy stage due to it's newfound scarcity, these eventually level off as dimishing returns to labour set in.

### Total output of the economy

```python
#Plot Total output
plt.clf()
plt.style.use('seaborn')
plt.plot(time_series_df['Total Output'], label = 'Total Output')
plt.axvline(27, lw=1, linestyle='dashed')
plt.xlabel('Year')
plt.ylabel('Output')
plt.title('Figure 4 - Total Output', fontweight='bold')
plt.legend()
plt.show()
```

**Figure 4** shows the dynamics of total output. We see growth in total output accelerating as capital is accumulated in the modern sector and labour is drawn in to exploit it. This continues up until the point where labour has fully migrated into the modern sector and the subsistence sector ceases to exist, shown by the dotted vertical line in **Figure 4**. At this point, the economy transitions to a one-sector capitalist model and exhibts growth dynamics in line with the standard **Solow** model. We see the leveling off of total output as the economy approaches the steady state level of capital and output as discussed previously.
